## 1.1 What is Python?
Python is one of the most widely used programming languages in the fields of robotics and artificial intelligence.
It is defined as *an interpreted, object-oriented, high-level programming language with dynamic semantics*. This implies the following:
- **Interpreted**: It means the language is not compiled. That is, a Python code can be executed directly. This also implies that errors are not detected until runtime.
- **Object-oriented**: Everything in Python is represented by objects, so classes and methods can be used. This allows for an organized and logical structure when programming.
- **High-level**: It allows programming using "common" words, achieving a functional program in a short time and without much complexity.
- **Dynamic semantics**: It implies that variables are objects, so it is not necessary to specify the data type. Python can determine that information.


```
# Example of dynamic semantics
x = 5   # Python recognizes that x contains a numeric value
x = x - 2 # The INT class allows mathematical operations
print(f'{x=}')
x = 'Hello' # Python recognizes that x now contains a string
x = x.upper() # The upper method belongs to the string class and is used to convert a string to uppercase
print(f'{x=}')
```

## **2. Tokenization and Embeddings with LLM**

Embeddings are mathematical representations (vectors) that models use to understand and work with words, phrases, or even documents.

They are a natural language processing technique that converts human language into mathematical vectors. They allow words to be treated as data and manipulated mathematically. This technique is widely used in artificial intelligence for tasks such as sentiment analysis, text classification, and machine translation.

### *How do embeddings work?*

An embedding converts text to numbers by following these steps:

1.   **Tokenization**: Before converting words into vectors, the text is divided into tokens. Tokens are like small pieces of text that the model can process.

    A token can be a word, part of a word, or even a character, depending on the model.
    
    *Example*: "Hello, world" can be tokenized as ["Hello", ",", "world"].
  

2.   **Vector Assignment**: Each token is assigned a vector of numbers. This vector has many dimensions and represents relationships and characteristics of the token in a mathematical space.

    (Example: [3.1, -1.2, 0.5, ...]).


3.   **Embedding Training**: Models learn to adjust the vectors during training to capture the meaning of the words.



There are certain models, generally the simpler ones, in which it is necessary to preprocess the text to reduce noise and simplify the text for more efficient analysis.

**Example applications:**
*   Sentiment analysis
*   Text classification

**Types of preprocessing:**
*   Stopword removal: Reduces words without semantic value (e.g., "and", "the", "a").
*   Lemmatization or stemming: Reduces words to their root form (e.g., "running" → "run").
*   Lowercase conversion: Avoids treating "House" and "house" as different terms.
*   Removal of punctuation or special characters: To ignore irrelevant elements in the task.

**Disadvantages:**
*   Loss of information: When removing stopwords or special characters, you can lose important context.
*   In modern models like GPT or BERT, embeddings already capture meaning and contextual relationships. Preprocessing text can remove information that the model needs to better understand. Example: "we are working" and "we work" have different nuances that the model could capture, but not if lemmatization is done.
*   Increased processing time: Some preprocessing techniques (like lemmatization) can be slow, especially with long texts.

This is why in complex models like GPT or BERT, preprocessing is not used as these models handle noise, stopwords, and special characters automatically.


OpenAI tokenization tool: https://platform.openai.com/tokenizer

### **Example with embeddingss**

```
from transformers import AutoTokenizer, AutoModel

# Selection of the pre-trained model
model_name = "distilbert-base-uncased"

# Download the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

print("Model and tokenizer ready.")
```

- Text preprocessing

```
# Import the necessary libraries
import re
import nltk
from nltk.corpus import stopwords

# Download the stopwords
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

# Preprocessing function
def preprocess_text(text):
    # Convert to lowercase and remove special characters
    text = re.sub(r"\W", " ", text.lower())
    # Remove stopwords
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text

# Example text
example_text = "This is an example of text preprocessing for the course of 2025"
processed_text = preprocess_text(example_text)

print("Original text:", example_text)
print("Preprocessed text:", processed_text)
```

- Tokenization and embeddings

```
# Example of tokenization
text = "In this course we are going to learn about AI models"
tokens = tokenizer(text, return_tensors="pt")

print("Generated tokens:")
print(tokens)

# Generation of embeddings
outputs = model(**tokens)
embeddings = outputs.last_hidden_state

print("\nDimensions of embeddings:", embeddings.shape)
```

**Tensorflow**

```
# Use of TensorFlow and PyTorch with Transformers
import torch
from transformers import TFAutoModel, AutoTokenizer, AutoModel
```

```
# Load the model with TensorFlow
print("\n--- Use of TensorFlow ---")
model_name = "bert-base-uncased"
tf_model = TFAutoModel.from_pretrained(model_name)
tf_tokenizer = AutoTokenizer.from_pretrained(model_name)

text = "Deep learning models are fascinating!"
tokens_tf = tf_tokenizer(text, return_tensors="tf")
outputs_tf = tf_model(**tokens_tf)

print("TensorFlow model outputs (shape):", outputs_tf.last_hidden_state.shape)
```

**Pytorch**

```
# Load the model with PyTorch
print("\n--- Use of PyTorch ---")
torch_model = AutoModel.from_pretrained(model_name)
tokens_torch = tf_tokenizer(text, return_tensors="pt")
outputs_torch = torch_model(**tokens_torch)

print("PyTorch model outputs (shape):", outputs_torch.last_hidden_state.shape)
```

### **Practical Activity**

```
# Text to preprocess and generate embeddings
text_to_process = """
Large language models like GPT and BERT have revolutionized natural language processing.
"""

# Steps:
# 1. Preprocess the text
# 2. Tokenize the text
# 3. Generate the embeddings

# Write your solution here:
processed_text = preprocess_text(text_to_process)
tokens = tokenizer(processed_text, return_tensors="pt")
outputs = model(**tokens)
embeddings = outputs.last_hidden_state

# Print the final dimensions of the embeddings
print("Final dimensions of embeddings:", embeddings.shape)
```
